```python
BASE_URL = "https://cloud.meters.taipit.ru"
API_URL = f"{BASE_URL}/api"
```

```python
s = requests.Session()
```

```python
# Request access token
print("Request access token")
payload = {
    'grant_type': 'password',
    'username': username,
    'password': password,
    'client_id': client_id,
    'client_secret': client_secret
}
url = f"{BASE_URL}/oauth/v2/token"
r = s.post(url, data=payload)
print(f'URL: {r.url}')
print(f'Payload: {payload}')
print(f'Status: {r.status_code}')
token = r.json()
print('Response:')
pprint(token)
token["expires_in"] = int(token["expires_in"])
token["expires_at"] = time.time() + token["expires_in"]
dt = datetime.fromtimestamp(token["expires_at"])
pprint(f"Token expired at {dt}")

```

```python
# Request refresh access token

print("Request refresh access token")
payload = {
    'grant_type': 'refresh_token',
    'refresh_token': token['refresh_token'],
    'client_id': client_id,
    'client_secret': client_secret
}
url = f"{BASE_URL}/oauth/v2/token"
r = s.post(url, data=payload)
print(f'URL: {r.url}')
print(f'Status: {r.status_code}')
new_token = r.json()
print('Response:')
pprint(new_token)
new_token["expires_in"] = int(new_token["expires_in"])
new_token["expires_at"] = time.time() + new_token["expires_in"]
dt = datetime.fromtimestamp(new_token["expires_at"])
pprint(f"Token expired at {dt}")
```

```python
auth = {"Authorization": f"Bearer {token['access_token']}"}
```

```python
# Settings
print('Settings')
sections = ['regions', 'meterTypes', 'controllers']
params = {'sections': ','.join(sections)}
url = f"{API_URL}/config/settings"
r = s.get(url, params=params, headers=auth)
print(f'URL: {r.url}')
print(f'Status: {r.status_code}')
settings = r.json()
print('Response:')
pprint(settings)
```

```python
# Warnings
print('Warnings')
params = {'action': 'getEntries'}
url = f"{API_URL}/warnings/list"
r = s.get(url, params=params, headers=auth)
print(f'URL: {r.url}')
print(f'Status: {r.status_code}')
warnings = r.json()
print('Response:')
pprint(warnings)
```
